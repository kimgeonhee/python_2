```python
# 2. 다형성에 대해 설명하고 다형성을 보여주는 자신만의 파이썬 코드 예제를 제시하시오.

#다형성이란 상속 관계 내의 다른 클래스들의 인스턴스들이 같은 멤버 함수 호출에 대해 각각 다르게 반응하도록 하는 기능

class Department:
    def slogan(self):
        print 'good!'
        
class Computer(Department):
    def slogan(self):
        print 'fighting Computer!'
        
class Mechatronics(Department):
    def slogan(self):
        print 'fighting Meca'
        
class Architecture(Department):
    def slogan(self):
        print 'fighting Architecture'
        
class Design(Department):
    def slogan(self):
        print 'fighting Design'

for x in (Computer(), Mechatronics(), Architecture(),Design()):
    x.slogan()

```

다형성이란 상속 관계 내의 다른 클래스들의 인스턴스들이 같은 멤버 함수 호출에 대해 각각 다르게 반응하도록 하는 기능이며 코드를 보면 slogan()이라는 멤버 함수를 통해 모든 클래스들의 각자 다르게 구현된 slogan()을 호출 할수 있습니다. 즉 같은 이름의 함수더라도 그 기능은 다르게 구현할수 있습니다.

```python
'''3.다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 
정의 코드를 제시하면 된다.)

•요구사항 1. 생성자에 count 값과 step 값을 인자로 받을 수 있다.
◾count: Counter 인스턴스가 지니는 초기 정수 값
◾step: Counter 인스턴스의 count가 증가되는 증분 (defalt 값: 1)

'''
class Counter:
    def __init__(self,count,step=1):
        self.count=count
        self.step=step
    def incr(self):
         self.count+=self.step
    def __str__(self):
        return '[Count(step:%s)]%s'%(self.step,self.count)
    def __call__(self):
        return self.incr()
    def __add__(self,other):
        return Counter(self.count+other)
    def __sub__(self,other):
        return Counter(self.count-other,2)
    def __cmp__(self,other):
        return self.count-other
c=Counter(10)
d=Counter(10,2)
print c
print d
c.incr()
d.incr()
print c
print d
c()
d()
print c
print d
c=c+5
d=d-5
print c
print d
print c>10
print d>10
print c<10
print d<10
print c==10
print d!=9

```

생성자 함수 __init__가 인자를 받을 수 있도록 하였고, step의 default 값을 1로 놓으라고 하였으므로 step=1을 하였습니다.  print를 위한 __str__ 함수와 count 값 증가를 위한 incr 함수는 count값과 step값을 더했고, 호출 가능형으로 만들기 위한 call 함수 및 연산을 위한 add, sub(2번째 step임을 나타내기 위해 step값에 2를 넣어주었습니다.) 함수와 비교를 위한 cmp 함수를 정의하였습니다.

```python
'''4.다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서
__init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.
•MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.
'''
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
```

먼저 init()함수는 생성자함수로서 객체에 인자로 받은 리스트의 모든 원소를 for문을 통해 하나하나 추가합니다. 그리고 최종적으로 eliminate_duplicate() 함수가 리스트를 호출 하도록 합니다. 여기서 eliminate_duplicate() 함수는 새로운 s리스트를 만들어 원소들을 하나씩 저장하지만 검사를 하여 중복되지 않는 것만 저장합니다. 이후 str()은 어떤식으로 print 될것인가를 정하는데 먼저 result에  MySet: {를 저장하고 그뒤에 받은 리스트의 원소들을 하나하나 추가 해줍니다. 그리고 마지막 띄어쓰기와 ,를 없애주고 }를 추가해준 result값을 리턴하여줍니다.

```python
'''5. 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오
•정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.

•요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.
>>> u = s | t 
 >>> print u 
 MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9} 

•요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.
>>> u = s & t 
 >>> print u 
 MySet: {2, 3} 

•요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.
>>> s = MySet([1, 2, 3]) 
 >>> t = MySet([3, 4, 5]) 
 >>> u = s - t 
 >>> print u 
 MySet: {1} 
'''
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
            
    #-------------------
    def union(self, A):  #같은 값이 없을 때만 리스트에 값을 추가해줍니다.
        for x in A:
            if x not in self:
                self.append(x)
        return MySet(self)
    def intersection(self, B): # 자신의 리스트와 다른 인스턴스의 리스트의 값 둘다 있는 원소만 새로운 w리스트에 추가해줍니다
        w=[]
        for x in self:
            if x in B:
                w.append(x)
        return MySet(w)
    def difference(self, C): #자신의 리스트중 다른 인스턴스의 리스트와 겹치지 않은 원소만 w에 추가해 줍니다.
        w=[]
        for x in self:
            if x not in C:  
                w.append(x)
        return MySet(w)
    def __or__(self,D):    # '|'연산자를 사용하기 위해 해사용 ,인자 값으로 자신과 다른 인스턴스를 받아 nion()함수값으로 반환해줍니다.
        return self.union(D)
    def __sub__(self,E):  #‘-’에 대한 연산자 오버로딩을 정의 합니다. 인자를 받아 difference()함수값으로 출력합니다
        return self.difference(E)
    def __and__(self,F):  #‘&’에 대한 연산자 오버로딩을 정의 합니다. 인자를 받아 intersection()함수값을 반환 해줍니다.
        return self.intersection(F)
    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
    

u=s|t
print u

s = MySet([1, 2, 2, 3])
t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])

u=s & t
print u

s = MySet([1, 2, 3]) 
t = MySet([3, 4, 5]) 
u = s - t 
print u 
```

```python
'''6. 5번 문제에서 정의한 MySet 클래스에 대해 다음 예제를 수행하면 오류없이 올바르게 동작하는 것을 확인할 수 있다.
다음 예제 내에 있는 len(), bool() 내장함수와 in 키워드 사용 예제가 별다른 메소드 정의를 하지 않았는 데도 올바르게 수행되는 이유를 설명하시오.
> \>\>\> s = MySet([1, 2, 3, 4, 5, 6])  <br/>
> \>\>\> print len(s)  <br/>
> 6 <br/>
> \>\>\> print bool(s)  <br/>
> True <br/>
> \>\>\> print 2 in s  <br/>
> True <br/>
'''

class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
            
    #-------------------
    def union(self, A):  #같은 값이 없을 때만 리스트에 값을 추가해줍니다.
        for x in A:
            if x not in self:
                self.append(x)
        return MySet(self)
    def intersection(self, B): # 자신의 리스트와 다른 인스턴스의 리스트의 값 둘다 있는 원소만 새로운 w리스트에 추가해줍니다
        w=[]
        for x in self:
            if x in B:
                w.append(x)
        return MySet(w)
    def difference(self, C): #자신의 리스트중 다른 인스턴스의 리스트와 겹치지 않은 원소만 w에 추가해 줍니다.
        w=[]
        for x in self:
            if x not in C:  
                w.append(x)
        return MySet(w)
    def __or__(self,D):    # '|'연산자를 사용하기 위해 해사용 ,인자 값으로 자신과 다른 인스턴스를 받아 nion()함수값으로 반환해줍니다.
        return self.union(D)
    def __sub__(self,E):  #‘-’에 대한 연산자 오버로딩을 정의 합니다. 인자를 받아 difference()함수값으로 출력합니다
        return self.difference(E)
    def __and__(self,F):  #‘&’에 대한 연산자 오버로딩을 정의 합니다. 인자를 받아 intersection()함수값을 반환 해줍니다.
        return self.intersection(F)
    
s = MySet([1, 2, 3, 4, 5, 6])  
print len(s)
print bool(s)
print 2 in s
```

```python
'''8. 1부터 5까지의 숫자를 영어로 쓰면 one, two, three, four, five 이고,
각 단어의 길이를 더하면 3 + 3 + 5 + 4 + 4 = 19 이므로 사용된 글자는 모두 19개입니다.

1부터 1,000까지 영어로 썼을 때는 모두 몇 개의 글자를 사용해야 할까요?

참고: 빈 칸이나 하이픈('-')은 셈에서 제외하며, 단어 사이의 and 는 셈에 넣습니다.
   예를 들어 342를 영어로 쓰면 three hundred and forty-two 가 되어서 23 글자,
   115 = one hundred and fifteen 의 경우에는 20 글자가 됩니다.
'''
def number(N):
    if N == 1000:
        return 'onethousand'
    elif N >= 100:
        if N % 100 == 0:
            return number(N/100)+'hundred'
        else:            
            return number(N/100)+'hundredand' + number(N%100)
    elif N >= 20:    
        return ['twenty', 'thirty', 'fourty', 'fifty', 'sixty', 'seventy', 'eighty', 'ninty'][N/10-2]+number(N%10)
    else:
        return ['', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine', 'ten', 'eleven', 'twelve', 'thirteen', 'fourteen', 'fifteen', 'sixteen', 'seventeen', 'eighteen', 'nineteen'][N]


print sum([len(number(n)) for n in range(1, 1001)])
```

```python
'''12. n의 약수들 중에서 자신을 제외한 것의 합을 d(n)으로 정의했을 때,
서로 다른 두 정수 a, b에 대하여 d(a) = b 이고 d(b) = a 이면 
a, b는 친화쌍이라 하고 a와 b를 각각 친화수(우애수)라고 합니다. 

예를 들어 220의 약수는 자신을 제외하면 1, 2, 4, 5, 10, 11, 20, 22, 44, 55, 110 이므로 그 합은 d(220) = 284 입니다.
또 284의 약수는 자신을 제외하면 1, 2, 4, 71, 142 이므로 d(284) = 220 입니다.
따라서 220과 284는 친화쌍이 됩니다. 

10000 이하의 친화수들을 모두 찾아서 그 합을 구하세요.
'''


def sum_y(number):
    
    result = 0
        
    for x in xrange(1,number+1):
        if number%x == 0:
            result += x
    return result
            
    


def findnum(number):
        x = sum_y(number) - number
        if sum_y(x) - x == number:
            return x
        return None

s=[]

for i in xrange(1, 10001):
    x = findnum(i)
    if (x != None) and (x != i) :
        s.append(x)
       
        
print sum(s)
```

findnum()함수에서 인자값의 약수를 sum_y()을 사용해 모두 더해준뒤 자기자신의 값을 빼서 x에 저장해 줍니다. 만약에 그 x값을 똑같이 sum_y()에 인자값으로 넣고 자신 x값을 뺀 값이 처음 넣은 값과 같다면 이는 친화수 이므로 x값을 반환하는 함수입니다.
